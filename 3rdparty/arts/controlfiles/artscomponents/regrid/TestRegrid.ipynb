```python
%matplotlib inline
import matplotlib.pyplot as plt
from PyARTS import artsXML
```

```python
gf_data = artsXML.load("gf_data.xml")
```

```python
def plot_orig_vs_regridded(gf_data, gf_regridded, name):
    plt.plot(gf_data['Pressure'], gf_data['Tensor3'].flatten(), color='#0000CC', label="original_data")
    plt.plot(gf_regridded['Pressure'], gf_regridded['Data'].flatten(), color='#CC0000', label=name)
    plt.xlim(0, 20000)
    plt.legend()
```

```python
for fname in ('none', 'both_sides', 'top', 'bottom'):
    fig = plt.figure()
    plot_orig_vs_regridded(gf_data, artsXML.load('gfREFERENCE_0p_' + fname + '.xml'), "GriddedFieldPRegrid zeropadding " + fname)
```

```python
for fname in ('none', 'both_sides', 'top', 'bottom'):
    fig = plt.figure()
    plot_orig_vs_regridded(gf_data, artsXML.load('gfREFERENCE_z_0p_' + fname + '.xml'), "GriddedFieldZToPRegrid zeropadding " + fname)
```
